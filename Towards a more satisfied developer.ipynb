```python
# Question 3.2: What techniques are most welcome among developers
    ## In general, newer techniques lead to better job and career satisfactory among developers
    ## Newer platforms such as AWS, Docker, iOS, Kubernetes, Microsoft Azure, Slack are more welcome than more traditional ones such as Windows, Android, Arduino, Linux, Raspberry Pi
    ## Newer databases such as Elasticsearch, DynamoDB, Redis are more welcome than old-fashioned Firebase, MySQL, Oracle
    ## Newer languages such as Clojure, Elixir, Go, Ruby, Scala, TypeScript are more welcome than older ones like C, VBA, PHP, Assembly
    ## Developing environment does not make much difference because they are much easier to evolve according to developers' needs
```

```python
def contain_words(words):
    '''
    INPUT
        words - a string of True, False or NaN from the data series
    
    OUTPUT
        return 1 if the string is  "True"
        return 0 otherwise
    
    '''
    if words == True:
        return 1
    else:
        return 0
```

```python
def calc_sat_score(df, category, technique, sat_type):
    '''
    INPUT
        df - the datafram to work with
        category - 4 categories: PlatformWorkedWith, DatabaseWorkedWith, DevEnviron, and LanguageWorkedWith
        technique - the technique choices under each category, in the form of '***' string
        sat_type - the type of satisfaction score: jobSatScore and careerSatScore
        
    OUTPUT
        return the average score for calculated technique
    '''
    
    # 1. Determine whether the column contains chosen technique
    contain_tech = 'contain_' + technique
    df[contain_tech] = df[category].str.contains(pat = technique)
    
    # 2. Convert [contain_tech] column into 1s and 0s based on the technique's existence
    yes_or_no = 'YorN_' + technique
    df[yes_or_no] = df[contain_tech].apply(contain_words)
    
    # 3. Calculate the average satisfaction score for the technique
    value = df[yes_or_no]*df[sat_type]
    score = value.sum()/(df[yes_or_no].sum())

    return score
```

```python
# Q 3.2.1, compute the most satisfactory techniques associated with platforms
platform_scores = pd.DataFrame(columns = ['platform_tech','jobSat_score','careerSat_score'])
platform_tech = ["Android", "Arduino", "AWS", "Docker", "Google Cloud Platform", "Heroku", "IBM Cloud or Watson", "iOS", "Kubernetes", "Linux", "Mac OS", "Microsoft Azure", "Raspberry Pi", "Slack", "Windows", "Wordpress"]
for tech in platform_tech:
    jobSat_result = calc_sat_score(df, 'PlatformWorkedWith', tech, 'jobSatScore')
    careerSat_result = calc_sat_score(df, 'PlatformWorkedWith', tech, 'careerSatScore')
    platform_scores = platform_scores.append({'platform_tech':tech, 'jobSat_score':jobSat_result, 'careerSat_score':careerSat_result}, ignore_index=True)
    
```
